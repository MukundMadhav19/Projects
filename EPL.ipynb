```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("C:\\Users\\Mukund\\Downloads\\Premier League 2024-25 stats.csv")
df.head()
```

```python
position_map = {
    'FW': 'Forward', 'FWD': 'Forward', 'ST': 'Forward',
    'MF': 'Midfielder', 'MID': 'Midfielder', 'CM': 'Midfielder', 'AM': 'Midfielder', 'DM': 'Midfielder',
    'DF': 'Defender', 'DEF': 'Defender', 'CB': 'Defender', 'LB': 'Defender', 'RB': 'Defender',
    'GK': 'Goalkeeper', 'Goalie': 'Goalkeeper', 'GKP' : 'Goalkeeper'
}

df['Position'] = df['Position'].replace(position_map)
```

```python
percent_cols = ['Conversion %', 'Crosses %', 'Passes%', 'fThird Passes %', 'gDuels %', 'aDuels %', 'Saves %']

for col in percent_cols:
    df[col] = df[col].str.rstrip('%').astype(float)

df.head()
```

```python
numeric_cols = df.columns.difference(['Player Name', 'Club', 'Nationality', 'Position'])  # exclude non-numeric
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.head()
```

```python
df['Goal Contributions'] = df['Goals']+df['Assists']
```

```python
df['Shot Accuracy'] = (df['Shots On Target'] / df['Shots'])
df.head()
```

```python
df.columns 
```

```python
df['Fair Play'] = (df['Yellow Cards'] + df['Red Cards'])/df['Appearances']
```

```python
df['Defensiveness'] = (df['Possession Won'] + df['Clearances'] + df['Interceptions'] + df['Tackles'] + df['gDuels Won'] + df['aDuels Won']+df['Blocks']) 
```

```python
df['Ariel Threat'] = (df['aDuels Won']/df['Aerial Duels'])
```

```python
df.drop('aDuels %',axis=1)
```

```python
df.drop('Saves %',axis=1)
```

```python
df['Saves per conceded'] = df['Saves']/df['Goals Conceded']
df.head()
```

```python
df.isnull().sum()
```

```python
df['Saves per conceded'] = df['Saves per conceded'].fillna(0)
```

```python
df['Shot Accuracy'] = df['Shot Accuracy'].fillna(0)
```

```python
df['Ariel Threat'] = df['Ariel Threat'].fillna(0)
```

```python
df.head()
```

```python
df['player_id'] = df.reset_index().index + 1
```

```python
cols = ['player_id'] + [col for col in df.columns if col != 'player_id']
df = df[cols]
df.columns
```

```python
player_features = ['Goals' , 'Assists', 'Shots', 'Shot Accuracy', 'Goal Contributions', 'Defensiveness', 'Saves per conceded', 'Ariel Threat', 'Passes%']
```

```python
X = df[player_features].fillna(0)
X.head()
```

```python
X.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_filled = X.fillna(0)  # or use mean()
X_scaled = StandardScaler().fit_transform(X_filled)
similarity_matrix = cosine_similarity(X_scaled)
```

```python
similar_names = []
similar_scores = []

for i, row in enumerate(similarity_matrix):
    row[i] = -1  # exclude self
    top_idx = np.argmax(row)
    similar_names.append(df.iloc[top_idx]['Player Name'])
    similar_scores.append(row[top_idx])
    
df['Most Similar Player'] = similar_names
df['Similarity Score'] = similar_scores
```

```python
top_n = 3

similar_df = pd.DataFrame({
    'player_id': df['player_id'],
    'Player Name': df['Player Name']
})

for n in range(1, top_n + 1):
    names = []
    scores = []
    for i, row in enumerate(similarity_matrix):
        row[i] = -1  # exclude self
        top_idxs = np.argsort(row)[-top_n:][::-1]  # top N indices in descending order
        names.append(df.iloc[top_idxs[n - 1]]['Player Name'])
        scores.append(row[top_idxs[n - 1]])
    similar_df[f'Similar Player {n}'] = names
    similar_df[f'Similarity Score {n}'] = scores
```
